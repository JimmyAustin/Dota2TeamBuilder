```python
import csv

def ready_line(line):
    line = {k.strip(): v.strip() for k, v in line.items()}
    line['HERO'] = line['HERO'].split('icon.png')[1].strip()
    line['A'] = line['A'].split(' ')[0].strip()
    return line
    
heroes = [ready_line(line) for line in csv.DictReader(open('./hero_stat.csv', 'r'))]
print(heroes)
```

```python
import json

hero_list = json.load(open('./heroes.json'))

hero_list_by_name = {x['localized_name']: x for x in hero_list['heroes']}
```

```python
for hero in heroes:
    hero_list_obj = hero_list_by_name[hero['HERO']]
    hero['ID'] = hero_list_obj['id']
    hero['NAME'] = hero_list_obj['name']
```

```python
with open('./stat_augmented_hero_list.json', 'wb') as fileref:
    fileref.write(json.dumps(heroes, indent=4).encode('utf8'))
```
